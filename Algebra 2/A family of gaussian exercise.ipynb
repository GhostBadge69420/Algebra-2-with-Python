```python
h = .999

t = np.linspace(-3,3,1001)

# gaussian
g = np.exp(-4*np.log(2)*t**2/h**2)


# plot
plt.plot(t,g)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Gaussian with FWMH=' + str(h))

plt.show()
```

```python
N = 100

G = np.zeros((N,len(t)))
h = np.zeros(N)

for i in range(N):
    h[i] = (i+1)/N
    G[i,:] = np.exp(-4*np.log(2)*t**2 / h[i]**2)

plt.pcolormesh(t,h,G)
plt.xlabel = ('t')
plt.ylabel = ('h')
plt.show()
```
