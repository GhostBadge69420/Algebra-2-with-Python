# Exercise

a = {1,2,3,4,5}

1) Σa/Σa^2 = 1/Σa

2) Πa/Πa^2 = 1/Πa

```python
#1)
a = np.arange(1,6)

suma = np.sum(a)
suma2 = np.sum(a**2)

ans1 = suma/suma2
ans2 = 1/suma

print('Option 1 = %s \n Option 2 = %s' %(ans1,ans2))
```

```python
#2)

proda = np.prod(a)
prod2 = np.prod(a**2)

ans1 = proda/prod2
ans2 = 1/proda

print('Option 1 = %s \n Option 2 = %s' %(ans1,ans2))
```
