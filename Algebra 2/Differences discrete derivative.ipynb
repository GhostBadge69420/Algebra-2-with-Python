f(x) = x^2

```python
x = np.linspace(-2,2,141)
f = x**2
df = np.diff(f)
dx = np.diff(x)

plt.plot(x,f,label='f')
plt.plot(x[:-1],df/dx,label='df/dx')

plt.axis([-2,2,-1,2])
plt.legend()
plt.grid()
plt.show()
```
