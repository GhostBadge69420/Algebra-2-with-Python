```python
z1 = complex(-3, 1)
z2 = complex(-1, 1)
sumCN = z1+z2
prodCN = z1*z2
 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
 
ax1.plot([0, np.real(z1)],[0, np.imag(z1)], 'b', label = 'z1')
ax1.plot([0, np.real(z2)],[0, np.imag(z2)], 'r', label = 'z2')
ax1.plot([0, np.real(sumCN)], [0, np.imag(sumCN)], 'g', label = 'z1+z2')
ax1.set(xlabel='Real', ylabel='Imag')
ax1.axis([-5, 5, -5, 5])
ax1.grid()
ax1.legend()
ax2.plot([0, np.real(z1)],[0, np.imag(z1)], 'b', label = 'z1')
ax2.plot([0, np.real(z2)],[0, np.imag(z2)], 'r', label = 'z2')
ax2.plot([0, np.real(prodCN)], [0, np.imag(prodCN)], 'g', label = 'z1 x z2')
ax2.set(xlabel='Real', ylabel='Imag')
ax2.axis([-5, 5, -5, 5])
ax2.grid()
ax2.legend()
```

```python
z1 = complex(-3,1)
z2 = complex(-1,1)

z1pz2 = z1+z2

plt.plot([0,np.real(z1)],[0,np.imag(z1)],label='z1')
plt.plot([0,np.real(z2)],[0,np.imag(z2)],label='z2')
plt.plot([0,np.real(z1pz2)],[0,np.imag(z1pz2)],label='z1+z2')

plt.xlabel('real')
plt.ylabel('imag')
plt.grid()
plt.axis('square')
plt.legend()
plt.axis([-5,5,-5,5])

plt.show()
```

```python
z1tz2 = z1*z2

plt.plot([0,np.real(z1)],[0,np.imag(z1)],label='z1')
plt.plot([0,np.real(z2)],[0,np.imag(z2)],label='z2')
plt.plot([0,np.real(z1tz2)],[0,np.imag(z1tz2)],label='z1+z2')

plt.xlabel('real')
plt.ylabel('imag')
plt.grid()
plt.axis('square')
plt.legend()
plt.axis([-5,5,-5,5])

plt.show()
```
