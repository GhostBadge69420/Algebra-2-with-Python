```python
def quadeq(a,b,c):
    
    out0 = ( -b - sp.sqrt(b**2 - 4*a*c)) / (2*a)
    out1 = ( -b + sp.sqrt(b**2 - 4*a*c)) / (2*a)

    return[out0,out1]

print(quadeq(1,4,5))
```

```python
x = np.linspace(-6,10,20)

fx = 1/(1+np.exp(x))

fmaxidx = np.argmin(abs(fx-.5))

plt.plot(x,fx,'bo-')
plt.plot(x[fmaxidx],fx[fmaxidx],'rs')

plt.title('Sigmoid function')
plt.show()
```

```python
x = np.linspace(0,12*np.pi,200)


fx = -(np.cos(x) + x**(1/2))


peeks = find_peaks(-fx)


plt.plot(x,fx)
plt.plot(x[peeks[0]],fx[peeks[0]],'o')
plt.show()
```
