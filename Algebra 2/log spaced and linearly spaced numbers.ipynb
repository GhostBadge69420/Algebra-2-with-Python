# Exercise

show that a linear scale approaches a logarithmic scale as the boundaries approach each other

```python
a = 40
b = 45
n = 50

li = np.linspace(a,b,n)
lo = np.logspace(np.log10(a),np.log10(b),n)

plt.plot(li,li,label='linear')
plt.plot(li,lo,label='log')

plt.legend()
plt.axis('square')
plt.show()
```
